```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,10):
    browser.get(f'https://www.traveloka.com/en-id/activities/search?st=GEO&eid=102746&sort=TRENDING&theme=DAY_TOUR&funnel_source=Merchandising.AA.Category-DAY_TOUR-web-ID-LandingPage&funnel_id=48191&internal_source=true{index}')
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='css-1dbjc4n r-16y2uox r-1wbh5a2 r-1wtj0ep r-1e081e0 r-1yzf0co')
    
    for row in rows:
        nama_wisata = row.find(class_='css-4rbku5 css-901oao css-cens5h r-t1w4ow r-ubezar r-b88u0q r-135wba7 r-c8eef1 r-fdjqy7').text
        lokasi_wisata = row.find(class_='css-901oao css-bfa6kz r-1ud240a r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-13hce6t r-fdjqy7').text
        rating_wisata = row.find(class_='css-901oao r-t1w4ow r-1enofrn r-majxgm r-1cwl3u0 r-13hce6t r-1d4mawv r-fdjqy7')
        if rating_wisata:
            rating_wisata = rating_wisata.text
        review = row.find(class_='css-901oao css-bfa6kz r-1ud240a r-t1w4ow r-1enofrn r-majxgm r-1cwl3u0 r-fdjqy7')
        if review:
            review = review.text
        harga_asli = row.find(class_='css-901oao r-1ud240a r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7 r-142tt33')
        if harga_asli:
            harga_asli = harga_asli.text
        harga_diskon = row.find(class_='css-901oao r-t1w4ow r-ubezar r-b88u0q r-rjixqe r-fdjqy7').text
        df = {
            'Name': nama_wisata,
            'Location': lokasi_wisata,
            'Rating': rating_wisata,
            'Total Review': review,
            'Price': harga_asli,
            'After Dicount': harga_diskon
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
